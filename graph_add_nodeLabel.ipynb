```python
import networkx as nx
```

```python
g = nx.read_gexf("graph.gexf")
nx.draw(g)
```

```python
# node_label hinzufügen
for node in g.nodes(data=True):
    node[1]["node_label"] = node[1]["surname"] + " " + node[1]["birth_year"]

nx.write_gexf(g, "./graph_relabeled.gexf", encoding="utf-8")
```
